Isso pode ser explicado pela, segundo um documento publicado pela OCDE, chamado "Panorama Educativo: Indicadores da OCDE", em que se afirma que no ano de 2002 houve uma falta de professores entre 14 países que foram analisados. Essa falta, segundo o autor, correspondia a cerca de 12% dos postos de trabalho docentes do encino secundário.

Esse bloco consiste na análise de um dos anos para ser possível a comparação entre os grupos.

```python
#gráfico de dispersão
# Entre os anos 2000 e 2001 houve um crescimento em ambos os gráficos:

#escolha de um ano para analise:
educacao_1996 = educacao_OCDE.loc[:,[2001]]
empregados_1996 = empregados_OCDE.loc[:,[2001]]

#Junção das variáveis quantitativas
df1 = empregados_1996.join(educacao_1996, how='inner', lsuffix='_empregados', rsuffix='_educacao')
df1
```

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre as variáveis:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('Educação Primária completa vs empregabilidade (OCDE)')
    plt.xlabel('% Educação primária completa')
    plt.ylabel('Empregados entre 15-24 anos (% populacional)')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
f = reta(df1['2001_educacao'], df1['2001_empregados'], 'red')
```

A inclinação positiva da linha de tendência acima mostra que a taxa de coonclusão do ensino primário está relacionada com a empregabilidade dos países de forma crescente.

Esse bloco consiste na análise de um dos anos para ser possível a comparação entre os grupos.

```python
df = empregados_n_1996.join(educacao_n_1996, how='inner', lsuffix='_n_empregados', rsuffix='_n_educacao')
df
```

* Gráfico de dispersão:

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre variáveis')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('Educação Primária completa vs empregabilidade (OCDE)')
    plt.xlabel('% Primary Completion')
    plt.ylabel('Empregados entre 15-24 anos (% populacional)')
    
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
f = reta(df['2001_n_educacao'], df['2001_n_empregados'], 'green')
```

A inclinação negativa da reta indica que enquanto uma das variáveis cresce, a tendência da outra é que haja um descréscimo. Nesse caso, percebe-se que um aumento na taxa de educação implica em um descréscimo na taxa de empregabilidade, confirmando o gráfico "Análise dos dados (países fora da OCDE) . 

Todavia, deve-se comparar os valores que essa reta está representando com a linha de tendência associada ao gráfico de dispersão apenas com os países que pertencem a OCDE. Conforme será exposto a seguir.

# Comparação entre os grupos:

```python
def linha_tendencia(dados_x, dados_y, c, label):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    plt.scatter(dados_x, dados_y, c=c, alpha=0.8, label=label)
    plt.title('Educação Primária completa vs empregabilidade (OCDE)')
    plt.xlabel('% Educação primária completa')
    plt.ylabel('Empregados entre 15-24 anos (% populacional)')
    plt.legend()
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)
    plt.axis([90, 115, 20, 60])
    plt.tight_layout()
```

```python
f = linha_tendencia(df['2001_n_educacao'], df['2001_n_empregados'], 'green', label = "Países fora da OCDE")
f1 = linha_tendencia(df1['2001_educacao'], df1['2001_empregados'], 'red', label = 'Países da OCDE')
```

Nota-se de acordo com esse gráfico que, de fato, para os países que compõem a Organização para Cooperação e Desenvolvimento Econômico há um considerável associação entre o o aumento da taxa conclusão do ensino primário e o aumento da taxa de empregabilidade nos países.
